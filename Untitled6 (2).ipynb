```python
tickers = ["7203.T", "6758.T", "8035.T", "8306.T", "4502.T", "9983.T"]
ticker_names = {
    "7203.T": "トヨタ",
    "6758.T": "ソニー",
    "8035.T": "東京エレクトロン",
    "8306.T": "三菱UFJ",
    "4502.T": "武田薬品",
    "9983.T": "ファーストリテイリング"
}

close_df = pd.DataFrame()
for ticker in tickers:
    data = yf.download(ticker, start="2015-01-01", end="2026-03-31")
    close_df[ticker] = data["Close"]

returns = close_df.pct_change().dropna()
```

```python
def optimize_portfolio(returns, n_iter=50000, seed=42):
    mean_returns = returns.mean() * 252
    cov_matrix = returns.cov() * 252
    n = len(returns.columns)

    np.random.seed(seed)
    best_weights = None
    best_sharpe = -np.inf

    for _ in range(n_iter):
        weights = np.random.random(n)
        weights /= weights.sum()
        ret = np.dot(weights, mean_returns)
        risk = np.sqrt(weights.T @ cov_matrix @ weights)
        sharpe = ret / risk
        if sharpe > best_sharpe:
            best_sharpe = sharpe
            best_weights = weights

    return best_weights, best_sharpe
```

```python
def plot_frontier(returns):
    mean_returns = returns.mean() * 252
    cov_matrix = returns.cov() * 252
    n = len(returns.columns)
    results = np.zeros((3, 5000))

    for i in range(5000):
        weights = np.random.random(n)
        weights /= weights.sum()
        ret = np.dot(weights, mean_returns)
        risk = np.sqrt(weights.T @ cov_matrix @ weights)
        results[0, i] = risk
        results[1, i] = ret
        results[2, i] = ret / risk

    plt.figure(figsize=(10, 6))
    plt.scatter(results[0], results[1], c=results[2], cmap="viridis", alpha=0.5)
    plt.colorbar(label="シャープレシオ")
    plt.xlabel("リスク（年率）")
    plt.ylabel("リターン（年率）")
    plt.title("効率的フロンティア（2015〜2019）")
    plt.show()

plot_frontier(returns["2015-01-01":"2019-12-31"])
```

```python
# 訓練期間で最適化
best_weights, best_sharpe = optimize_portfolio(returns["2015-01-01":"2019-12-31"])

print("最適重み（2015〜2019）")
for ticker, w in zip(tickers, best_weights):
    print(f"{ticker_names[ticker]}: {w:.1%}")
print(f"シャープレシオ: {best_sharpe:.3f}")

# 検証期間で比較
returns_test = returns["2020-01-01":"2026-03-31"]
nikkei = yf.download("^N225", start="2020-01-01", end="2026-03-31")["Close"]
nikkei_returns = nikkei.pct_change().dropna()

portfolio_cumulative = (1 + (returns_test * best_weights).sum(axis=1)).cumprod()
nikkei_cumulative = (1 + nikkei_returns).cumprod()

plt.figure(figsize=(12, 6))
plt.plot(portfolio_cumulative, label="最適ポートフォリオ")
plt.plot(nikkei_cumulative, label="日経225")
plt.title("累積リターン比較（2020〜2026検証）")
plt.ylabel("累積リターン（倍）")
plt.legend()
plt.show()
```

```python
# 訓練期間でランダム1000通りの重みを生成・固定
np.random.seed(0)
n_random = 1000
random_weights = []

for _ in range(n_random):
    w = np.random.random(n)
    w /= w.sum()
    random_weights.append(w)

# 検証期間のリターン
returns_test = returns["2020-01-01":"2026-03-31"]
mean_test = returns_test.mean() * 252
cov_test = returns_test.cov() * 252

# ランダム1000通りのシャープレシオを検証期間で計算
random_sharpes = []
for w in random_weights:
    ret = np.dot(w, mean_test)
    risk = np.sqrt(w.T @ cov_test @ w)
    random_sharpes.append(ret / risk)

# 最適化ポートフォリオのシャープレシオを検証期間で計算
opt_ret = np.dot(best_weights, mean_test)
opt_risk = np.sqrt(best_weights.T @ cov_test @ best_weights)
opt_sharpe = opt_ret / opt_risk
# 結果表示
rank = sum(s < opt_sharpe for s in random_sharpes)
percentile = rank / n_random * 100

print(f"最適化シャープレシオ: {opt_sharpe:.3f}")
print(f"ランダムの中央値: {np.median(random_sharpes):.3f}")
print(f"上位{100 - percentile:.1f}%")

# ヒストグラム
plt.figure(figsize=(10, 6))
plt.hist(random_sharpes, bins=50, alpha=0.7, label="ランダム1000通り")
plt.axvline(opt_sharpe, color="red", linewidth=2, label=f"最適化（{opt_sharpe:.3f}）")
plt.xlabel("シャープレシオ（検証期間）")
plt.ylabel("頻度")
plt.title("最適化 vs ランダム1000通り（検証期間2020〜2026）")
plt.legend()
plt.show()
```
